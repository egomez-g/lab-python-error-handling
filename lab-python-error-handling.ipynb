````python
def getPrices(products):
    prices = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                prices[product] = int(input(f"Enter the price of {product}"))
                if prices[product] < 0:
                    raise ValueError ("Invalid price! Please enter a non-negative value.")
                valid_input = True
            except ValueError as error:
                print(f"Error: {error}")
    return prices
````

````python
def totalPrice(prices, costumer_orders):
    return sum([prices[order] for order in costumer_orders if order in prices])
````

````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory
````

````python
def updateInventory(costumer_orders, inventory):
    for item in costumer_orders:
        if item in inventory:
            inventory[item] -= 1

    to_remove = [i for i in inventory if inventory[i] == 0]

    inventory = {i: inventory[i] for i in inventory if i not in to_remove}

    return inventory
````

````python
def get_costumer_orders(inventory):
    valid_input = False
    while not valid_input:
        try:
            inp_str = int (input("Enter the number of customer orders: "))
            if inp_str <= 0:
                raise ValueError ("Invalid input! Please enter a value bigger than 0.")
            valid_input = True
        except ValueError as error:
            print(f"Error: {error}")

    customer_orders = set()
    for i in range(inp_str):
        valid_input = False
        while not valid_input:
            try:
                prod = (input("Enter the name of a product that a customer wants to order: "))
                if (prod not in inventory):
                    raise ValueError ("We dont have this product")
                elif (prod in customer_orders):
                    raise ValueError ("You already ordered this product")
                valid_input = True
            except ValueError as e:
                print(f"Error: {e}")
        customer_orders.add(prod)

    return customer_orders

````

````python

def main():
    products = {
        "t-shirt",
        "mug",
        "hat",
        "book",
        "keychain"
    }
    
    inventory = initialize_inventory(products)
    
    customer_orders = get_costumer_orders(inventory)

    print("Order statistics:")
    print(f"Total Products Ordered: {len(customer_orders)}")
    #percentage
    order_status = len(customer_orders) / len(products) * 100
    print(f"Percentage of Unique Products Ordered: {order_status}%\n")
    
    inventory = updateInventory(customer_orders, inventory)
    print("Updated inventory")
    for i in inventory:
        print(f"{i} : {inventory[i]}")
    prices = getPrices(customer_orders)
    print (f"Total Price: {totalPrice(prices, customer_orders)}")

if __name__ == "__main__":
    main()
````
